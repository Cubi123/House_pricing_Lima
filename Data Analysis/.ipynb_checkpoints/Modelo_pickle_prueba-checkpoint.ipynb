```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('data_para_preprocessing.csv')
data.head(2)
```

```python
columnas = ['Area_constr_m2','latitud','longitud','Area_total_m2','Distrito','Provincia','Dormitorios']
X = data[columnas]
y = data['Precio']
```

###  A partir de acá tengo que hacer la limpieza de data

```python
dists = list(X.Distrito.unique())
provs = list(X.Provincia.unique())
dict_provs = {v:k for (k,v) in [*enumerate(provs)]}
dict_dist = {v:k for (k,v) in [*enumerate(dists)]}

pickle.dump(dists, open('distritos_dict.pkl','wb'))
pickle.dump(provs, open('provincias_dict.pkl','wb'))
```

```python
X.Distrito = X.Distrito.map(dict_dist)
X.Provincia = X.Provincia.map(dict_provs)
```

```python
X.loc[(X.Area_total_m2=='NoEspecifica'),'Area_total_m2'] =  225.0
```

```python
X.Area_total_m2 = pd.to_numeric(X.Area_total_m2)
```

```python
X.Dormitorios = X.Dormitorios.replace({'5+':5})
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X, y)
```

```python
import pickle
```

```python
pickle.dump(lr_model, open('model.pkl','wb'))
```
